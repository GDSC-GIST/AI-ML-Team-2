Seaborn 그래프 정리 & 사용법
https://coding-kindergarten.tistory.com/127

```python
#Pclass, Survived기준 그래프만들기

ax = sns.countplot(data = titanic,
                   x = 'Survived',
                   hue = 'Pclass')

# 막대그래프 위에 인원수 표시하기
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height + 3, height, ha = 'center', size = 9)
```

```python
#Pclass안에 있는 값, 그에 해당하는 데이터 개수
titanic['Pclass'].value_counts()
```

```python
#Pclass에서 Survived 기준으로 분류하고, 그에 해당하는 데이터 개수
titanic.groupby(['Pclass', 'Survived']).size()
```

https://www.delftstack.com/ko/howto/python-pandas/how-to-count-the-frequency-a-value-occurs-in-pandas-dataframe/

```python
#퍼센트로 바꾸기

percent1 = [80 / (80 + 136), 136 / (80 + 136)]
percent2 = [97 / (97 + 87), 87 / (97 + 87)]
percent3 = [372 / (372 + 119), 119 / (372 + 119)]
surv = ['Dead', 'Survived']
```

```python
p1 = sns.barplot(x = surv,
                 y = percent1)
p1.set_title("Pclass1(%)")

for p in p1.patches:
    height = p.get_height()
    p1.text(p.get_x() + p.get_width() / 2., height + 0.02, '%.2f' % float(height), ha = 'center', size = 9)
```

```python
p2 = sns.barplot(x = surv,
                 y = percent2)
p2.set_title("Pclass2(%)")

for p in p2.patches:
    height = p.get_height()
    p2.text(p.get_x() + p.get_width() / 2., height + 0.02, '%.2f' % float(height), ha = 'center', size = 9)
```

```python
p3 = sns.barplot(x = surv,
                 y = percent3)
p3.set_title("Pclass3(%)")

for p in p3.patches:
    height = p.get_height()
    p3.text(p.get_x() + p.get_width() / 2., height + 0.02, '%.2f' % float(height), ha = 'center', size = 9)
```

나이와 생존 사이 관계 알아보기!
===

```python
titanic.head()
```

```python
#나이가 빈칸인 사람들은 제외하고 추출하기

condition = (titanic['Age'] >= 0) & (titanic['Age'] <= 120)

agedf = titanic.loc[condition, ['PassengerId', 'Age', 'Survived']]

agelist = agedf['Age'].values.tolist()
agelist
```

```python
print("Age maximum : ", np.max(agelist))
print("Age minimum : ", np.min(agelist))
```

```python
agedf.head()
```

```python
agedfsort = agedf.sort_values(by = 'Age', ascending = True)

agedfsort.head()
```

```python
condition = (agedf['Age'] <= 18)

youngage = agedf.loc[condition]
youngage.head()
```

```python
youngage.index
#139명
```

```python
youngage['Survived'].value_counts()
```

```python
percent = [69 / 139, 70 / 139]
surv = ['Dead', 'Survived']

age1 = sns.barplot(x = surv,
                   y = percent)
age1.set_title("Young (under 18)")

for p in age1.patches:
    height = p.get_height()
    age1.text(p.get_x() + p.get_width() / 2., height + 0.02, '%.4f' % float(height), ha = 'center', size = 9)
```
