```python
#make_csv_dataset 이용해서 데이터프레임 -> 텐서플로우에서 사용 가능한 텐서로 바꾸기 & dataset 형태(feature & label 모두 포함)로 만들기

#from_tensor_slices, convert_to_tensor를 사용하지 못하는 것 같아요! (정확하진 않지만, male/female, S/Q/C처럼 숫자가 아닌 데이터가 포함되어서 그런 걸로..)
#숫자가 아닌 데이터가 있기 때문에, np.array(), dict()로 형변환해서 바로 넘길수도 없어요!

titanic_train = tf.data.experimental.make_csv_dataset(
    'Train_ppc.csv',
    batch_size = 5,
    label_name = 'Survived',
    select_columns = ['Pclass', 'Age', 'Sex', 'SibSp', 'Parch', 'Embarked', 'Survived'],
    num_epochs = 1,
    ignore_errors = True)

#shuffle_buffer_size도 사용 가능!
```

```python
#Densefeature layer에 넣기 위한 feature column 배열 만들기
#위 블럭에서 만든 string형태 배열을 쓸 수 없고, 텐서플로우에서 지정한 형식으로 만들어야 하기 때문에 tf.feature_column.~~를 사용해요!

#범주형 데이터인 경우
Pclass = tf.feature_column.categorical_column_with_identity(
    key = 'Pclass',
    num_buckets = 3)

#범주형 데이터인 경우(categorical~~을 사용하는 경우) 위에 걸 그대로 못 쓰고, indicator / embedded로 한번 처리해줘야 한다고 해요!
Pclass_ind = tf.feature_column.indicator_column(Pclass)

# 연속형 데이터를 특정 기준점 기준 분류해서 범주형으로 만들기
Age = tf.feature_column.numeric_column("Age")
Age_bucket = tf.feature_column.bucketized_column(
    source_column=Age,
    boundaries = [18, 65])

# 범주형 데이터이고, 텍스트로 구분되는 경우
Sex = tf.feature_column.categorical_column_with_vocabulary_list(
    key = 'Sex',
    vocabulary_list=['male', 'female'])

Sex_ind = tf.feature_column.indicator_column(Sex)

SibSp = tf.feature_column.categorical_column_with_identity(
    key = 'SibSp',
    num_buckets = 7)

SibSp_ind = tf.feature_column.indicator_column(SibSp)

Parch = tf.feature_column.categorical_column_with_identity(
    key = 'Parch',
    num_buckets = 7)

Parch_ind = tf.feature_column.indicator_column(Parch)

Embarked = tf.feature_column.categorical_column_with_vocabulary_list(
    key = 'Embarked',
    vocabulary_list=['S', 'C', 'Q'])

Embarked_ind = tf.feature_column.indicator_column(Embarked)

feature_col = [Pclass_ind, Age_bucket, Sex_ind, SibSp_ind, Parch_ind, Embarked_ind]
```

```python
#Dense layer에 넣기 전에, 모델한테 우리 데이터의 feature가 무엇인지 알려주기 위해 DenseFeatures레이어 필수!

Densemodel = tf.keras.models.Sequential([
    
    tf.keras.layers.DenseFeatures(feature_col, trainable=True),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(64, activation = 'relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(2, activation = 'softmax')
    
])
```

https://stackoverflow.com/questions/60071304/valueerror-passing-a-dictionary-input-to-a-sequential-model-which-doesnt-have

```python
Densemodel.compile(optimizer = 'adam',
                   loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                  metrics = ['accuracy'])
```

```python
EPOCHS = 500

model_checkpoint = keras.callbacks.ModelCheckpoint(
"my_checkpoint.h5", save_best_only = True)

early_stopping = keras.callbacks.EarlyStopping(patience = 70)


#왜인지는 모르겠지만, 이 데이터를 쓸 때는 batch_size는 지정할 수 없다고 에러가 뜨기 때문에 batch_size는 없습니다
#Densemodel = Densemodel.fit이 아니라, 그냥 Densemodel.fit으로 쓰면 뒤에서 Densemodel.history['accuracy']실행할 때 에러 나요!
Densemodel.fit(titanic_train, validation_data = titanic_val,
              epochs = EPOCHS,
              callbacks = [early_stopping, model_checkpoint])
```
